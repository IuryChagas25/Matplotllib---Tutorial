```python
import matplotlib.pyplot as plt  #importando a matplotlib e a parte da pyplot
%matplotlib inline              
import numpy as np               #incluindo a Numpy e Pandas que será úteis no nosso processo (não é necessário conhecimento sobre elas no momento, irei explicar cada operação)
import pandas as pd

%reload_ext autoreload         
%autoreload 2 
```

**Vamor introduzir com uma Plotagem simples de um gráfico exponencial:**

```python
x = np.linspace(0,9,13,20)       #definindo um conjunto de numeros com  espaçamento entre eles
y = x**2                       #coordenada y recebe os valores de x e os coloca em uma exponencial de razão 2 
```

```python
plt.plot(x,y)                  #Definindo a plotagem pelo comando plt.plot(variavel1,variavel2), não necessariamente o nome delas deverá ser x e y
plt.title("Plotagem Simples")  #criando um título para a plotagem
plt.xlabel("Valores de X")     #Informando o que há no eixo x
plt.ylabel("Valores de Y")     #Informando o que há no eixo y
plt.show()                     #Exibindo a plotagem
```

**Vamos agora plotas vários gráficos de uma única vez, poderemos usar isso para comparar dados ou apenas como recurso de melhor visualização:**

```python
                               #Podemos utilizar os dados já conhecidos de x e y para plotar subplotagens em um intervalo especifico do gráfico
plt.subplot(1,2,1)             #Com a mesma base de valores no intervalo (1,2) é possível plotas diferentes gráficos
plt.plot(x,y,'r')              #a terminação 'r','g' e 'b' refere-se ao nome da cor para o gráfico
plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

**Vamos agora usar Figure Objects que é um tipo de gráfico mais sofisticado onde se pode comparar curvas sobre outras e detalhar melhor os dados:**

```python
figura = plt.figure(figsize = (5,4), dpi = 100)     #Definindo uma figura de plotagem com tamanho 5x4
eixo_1= figura.add_axes([0.1,0.1,1.0,1.0])          #Definindo valores de borda,altura e escala em porcentagem do figsize definido acima
eixo_1.set_xlabel('Eixo x')                         #Descrevendo os dados do eixo x
eixo_1.set_ylabel('Eixo y')                         #Descrevendo os dados do eixo y
eixo_1.set_title('Gráfico mais avançado')           #Definindo um título
eixo_1.plot(x,y,label = 'x/x^2')                    #pondo uma legenda na curva da exponencial de razão x/x^2
eixo_1.plot(y,x,label = 'x^2/x')                    #pondo uma legenda na curva da exponencial de razão x^2/x
eixo_1.legend(loc= 0 )                              #incluindo a localização da legenda (0 e 1 = superior direito, 2 = superior esquerdo, 3 = inferior esquerdo, 4 = inferior direito)
#--------------------------------------------------------------------------------------------------------------------------------#
eixo_2 = figura.add_axes([0.4,0.4,0.5,0.4])         #Definindo a localização percentual da imagem do gráfico do eixo 2
eixo_2.set_xlabel('Eixo x2')                        #Descrevendo eixo x                       
eixo_2.set_ylabel('Eixo y2')                        #Descrevendo eixo y
eixo_2.set_title('Gráfico de região aproximada')    #Definindo um título
eixo_2.plot(x,y,'g')                                #Plotando o gráfico na cor verde
```

**Sub plotagens em Figure Objects:**

```python
figura_2, eixo_2 = plt.subplots(figsize = (8,5), nrows = 1, ncols = 3)  #Definindo figura_2 e o eixo 2 como subplotagens. nrows = 1 * ncols => resulta em 3 subgráficos
plt.tight_layout()                                                      #caso nrows = 2 e n cols = 2 => 4 subgráficos.
#------------------------(x/x^2)------------------------#
eixo_2[0].set_title('SubPlotagem')                      #Definindo o titulo
eixo_2[0].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[0].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[0].plot(x,y,label = 'x/x^2')                     #plotando x/x^2
eixo_2[0].legend(loc = 4)                               #colocando a legenda no canto inferior direito
#------------------------(x^2/x)------------------------#
eixo_2[1].set_title('SubPlotagem')                      #Definindo o titulo                     
eixo_2[1].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[1].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[1].plot(y,x,color = 'orange',label = 'x^2/x')    #plotando x^2/x
eixo_2[1].legend(loc = 4)                               #colocando a legenda no canto inferior direito
#--------------------(x/x^2)&(x^2/x)--------------------#
eixo_2[2].set_title('SubPlotagem')                      #Definindo o titulo
eixo_2[2].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[2].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[2].plot(x,y,label = 'x/x^2')                     #plotando x/x^2
eixo_2[2].plot(y,x,label = 'x^2/x')                     #plotando x^2/x
eixo_2[2].legend(loc = 0)                               #colocando a legenda no canto superior direito
```

```python
figura_2.savefig('Subplotagensfig.png')   #salvando as plotagens feitas na figura 2
```
