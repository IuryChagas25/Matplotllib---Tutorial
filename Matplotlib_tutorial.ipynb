### Plotagem de funções
***
Vamor introduzir com uma Plotagem simples de um gráfico exponencial.

```python
#definindo um conjunto de numeros com  espaçamento entre eles
x = np.linspace(0,9,13,20) 
#coordenada y recebe os valores de x e os coloca em uma exponencial de razão 2 
y = x**2                       
```

```python
plt.figure(figsize = (10,5))
#Definindo a plotagem pelo comando plt.plot(variavel1,variavel2), não necessariamente o nome delas deverá ser x e y
plt.plot(x,y)  
#criando um título para a plotagem
plt.title("Plotagem Simples")  
 #Informando o que há no eixo x
plt.xlabel("Valores de X")  
#Informando o que há no eixo y
plt.ylabel("Valores de Y") 
 #Exibindo a plotagem
plt.show()                    
```

### Multi-gráficos em um mesmo ambiente de plotagem
***
Vamos agora plotas vários gráficos de uma única vez, poderemos usar isso para comparar dados ou apenas como recurso de melhor visualização.

```python
plt.figure(figsize = (15,8))
#Podemos utilizar os dados já conhecidos de x e y para plotar subplotagens em um intervalo especifico do gráfico 
#Com a mesma base de valores no intervalo (1,2) é possível plotas diferentes gráficos
plt.subplot(1,2,1)  
#a terminação 'r','g' e 'b' refere-se ao nome da cor para o gráfico
plt.plot(x,y,'r')   
#criando um título para a plotagem
plt.title("Plotagem Simples 1")  
 #Informando o que há no eixo x
plt.xlabel("Valores de X")  
#Informando o que há no eixo y
plt.ylabel("Valores de Y") 
plt.subplot(1,2,2)
plt.plot(x,y,'g')
#criando um título para a plotagem
plt.title("Plotagem Simples 2")  
 #Informando o que há no eixo x
plt.xlabel("Valores de X")  
#Informando o que há no eixo y
plt.ylabel("Valores de Y") 
```

### Plotagens sofisticadas
***
Vamos agora usar `Figure Objects` que é um tipo de gráfico mais sofisticado onde se pode comparar curvas sobre outras e detalhar melhor os dados.

```python
#Definindo uma figura de plotagem com tamanho 5x4
figura = plt.figure(figsize = (5,4), dpi = 100) 
#Definindo valores de borda,altura e escala em porcentagem do figsize definido acima
eixo_1= figura.add_axes([0.1,0.1,1.0,1.0])       
#Descrevendo os dados do eixo x
eixo_1.set_xlabel('Eixo x')      
#Descrevendo os dados do eixo y
eixo_1.set_ylabel('Eixo y')     
#Definindo um título
eixo_1.set_title('Gráfico mais avançado')     
#pondo uma legenda na curva da exponencial de razão x/x^2
eixo_1.plot(x,y,label = 'x/x^2')   
#pondo uma legenda na curva da exponencial de razão x^2/x
eixo_1.plot(y,x,label = 'x^2/x')  
#incluindo a localização da legenda (0 e 1 = superior direito, 2 = superior esquerdo, 
# 3 = inferior esquerdo, 4 = inferior direito)
eixo_1.legend(loc= 0 )                              
#-------------------------------------------------------------------------------------#
#Definindo a localização percentual da imagem do gráfico do eixo 2
eixo_2 = figura.add_axes([0.4,0.4,0.5,0.4])    
#Descrevendo eixo x    
eixo_2.set_xlabel('Eixo x2')             
#Descrevendo eixo y
eixo_2.set_ylabel('Eixo y2')    
#Definindo um título
eixo_2.set_title('Gráfico de região aproximada')   
#Plotando o gráfico na cor verde
eixo_2.plot(x,y,'g')                                
```

### Sub plotagens em Figure Objects
***

```python
#Definindo figura_2 e o eixo 2 como subplotagens. nrows = 1 * ncols => resulta em 3 subgráficos
#caso nrows = 2 e n cols = 2 => 4 subgráficos.
figura_2, eixo_2 = plt.subplots(figsize = (14,4), nrows = 1, ncols = 3)  
plt.tight_layout()                                                      
#------------------------(x/x^2)------------------------#
eixo_2[0].set_title('SubPlotagem 1')                      #Definindo o titulo
eixo_2[0].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[0].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[0].plot(x,y,label = 'x/x^2')                     #plotando x/x^2
eixo_2[0].legend(loc = 4)                               #colocando a legenda no canto inferior direito
#------------------------(x^2/x)------------------------#
eixo_2[1].set_title('SubPlotagem 2')                      #Definindo o titulo                     
eixo_2[1].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[1].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[1].plot(y,x,color = 'orange',label = 'x^2/x')    #plotando x^2/x
eixo_2[1].legend(loc = 4)                               #colocando a legenda no canto inferior direito
#--------------------(x/x^2)&(x^2/x)--------------------#
eixo_2[2].set_title('SubPlotagem 3')                      #Definindo o titulo
eixo_2[2].set_xlabel('Sub Eixo X')                      #Definindo o eixo x
eixo_2[2].set_ylabel('Sub Eixo Y')                      #Definindo o eixo y
eixo_2[2].plot(x,y,label = 'x/x^2')                     #plotando x/x^2
eixo_2[2].plot(y,x,label = 'x^2/x')                     #plotando x^2/x
eixo_2[2].legend(loc = 0)                               #colocando a legenda no canto superior direito
```

### **Histogramas**
*Certo, agora iremos implementar histogramas com base em uma geração de números aleatórios com a numpy*

```python
numb1 = np.random.randint(1,4,7000)                                 #gerando 7000 numeros entre 1 e 4
numb2 = np.random.randint(1,4,7000)                                 #gerando 7000 numeros entre 1 e 4
soma = numb1 + numb2                                                #somando os numeros sorteados
plt.hist(soma,bins = 5,density=True,stacked=True,color = 'orange',cumulative = True)  
#plotando um histograma dos numeros aleatórios
                                                                    #bins = pode ser definido como a precisão em barras, muda-se para melhor a visualização
                                                                    #cumulative = true -> será mostrado o gráfico com as maiores frequências de digito
                                                                    #color = cor
                                                                    #histtype ='step' -> será exibido apenas as linhas do gráfico sem as barras
```

```python
plt.hist(soma,bins = 5,density=True,stacked=True,color = 'orange',histtype='step',cumulative = True)
```

```python
plt.hist(soma,bins = 10,density=True,stacked=True,orientation='horizontal')  #também é possível mudar a orientação com orientation = 'horizontal/vertical'
```

### Histogramas de imagens
*sim, é possível plotar um histograma de uma imagem, ele será criado a partir do números de pixels (na escala rgb) que aparecerão na imagem*

```python
import cv2
```

```python
imagem = plt.imread('12.jpg')           #Importando e visualizando a imagem com a matplotlib
img = plt.imshow(imagem)
```

```python
#lendo imagem com o comando cv2.imread() importado da biblioteca Opencv
img = cv2.imread('12.jpg',0)
  
# encontrando a sequencia de pixel em um range de 0 a 255 (RGB)
histr = cv2.calcHist([img],[0],None,[256],[0,256])
  
# Mostrando a plotagem gráfica da imagem
plt.plot(histr)
plt.show()
```

```python
plt.hist(img.ravel(),256,[0,255])            #Também é possível observar a imagem dessa forma
plt.show()
```
